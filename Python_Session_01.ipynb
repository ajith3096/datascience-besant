# Datatypes
Python has the following data types built-in by default, in these categories:



There is no need to declare the datatype or assign the same. The datatype is automatically assigned ny python interpreter.
But, Python also has provisons to set specific data types.


```python
x = "Hello World"
type(x)
```

```python
x = 20
type(x)
```

```python
x = 20.5
type(x)
```

```python
x = 1j
type(x)
```

```python
x = ["apple", "banana", "cherry"]
type(x)
```

```python
x = ("apple", "banana", "cherry")
type(x)
```

```python
x = range(6)
type(x)
```

```python
x = {"name" : "John", "age" : 36}
type(x)
```

```python
x = {"apple", "banana", "cherry"}
type(x)
```

```python
x = frozenset({"apple", "banana", "cherry"})
type(x)
```

```python
x = True
type(x)
```

```python
x = bytearray(5)
type(x)
```

```python
x = memoryview(bytes(5))
type(x)
```

```python
x = None
type(x)
```
